# **Tuple in Python**
Tuple is an immutable collection of objects

**Creating Tuple**

Tuple in Python can be created by just placing the sequence inside the round brackets().

```python
Tuple1=("English","Marathi","Hindi","Science","Social Studies")
Tuple2=(1,2,3,4,5)
type(Tuple1)
```

**Tuple with single element**

Tuples can also be created with a single element, but there must be a trailing ‘comma’ to make it a tuple or else it may form str or int .

```python
Tuple3=('hello',)
type(Tuple3)
```

**Accessing Tuple Element**

Tuples are immutable, and usually, they contain a sequence of heterogeneous elements that are accessed via unpacking or indexing

```python
Tuple1[0] #0'th index element in list1
```

```python
#tuple unpacking
a, b, c, d, e= Tuple1
print(a)
print(b)
print(c)
```

**Reverse Indexing**

Reverse indexing in Python allows you to access elements from the end of a sequence using negative indices.

The last element has an index of -1, the second-to-last has an index of -2, and so on.

```python
print(Tuple1[-1])
print(Tuple2[-4])
```

**Tuple Slicing**

slicing is a technique for extracting a portion (or a subset) of elements of tuple

The general syntax is **tuple_name[start:stop:step]**


```python
print(Tuple1[2:4]) #print from 2 upto 4 (excluding)
print(Tuple2[::2]) #print from start to end with step 2
print(Tuple2[::-1]) #print reverse list
```

**Tuple Concatenation**

Concatenation of tuple is the process of joining two or more Tuples. Concatenation is done by the use of ‘+’ operator.

```python
Tuple3=Tuple1+Tuple2
print(Tuple3)
```

**Deleting a tuple**

Tuples are immutable and hence they do not allow deletion of a part of it. The entire tuple gets deleted by the use of del() method.

```python
del(Tuple3)
print(Tuple3)
```

**index() Method**

Returns index of first matched element

```python
list2=(1,2,3,2,1)
print(list2.index(1)) #returns first matched index of given element
print(list2.index(2))
```

**count() Method**

It returns the count of the number an items repeats which is passed as an argument

```python
Tuple1=(1,2,3,2,1)
Tuple1.count(2)
```

**all() Method**

Returns true if tuple is empty or all elemnts are true

```python
Tuple3=(True,True,True)
Tuple2=()
print(all(Tuple3))
print(all(Tuple2))
```

**any() Method**

return true if any element of the tuple is true. if tuple is empty, return false

```python
Tuple3=(True,False,True)
Tuple2=()
print(any(Tuple3))
print(any(Tuple2))
```

**sorted() Method**

Returns a sorted list

```python
Tuple1=(5,2,8,1,4)
print(sorted(Tuple1))
```

**Min Max Sum Methods**

**Difference of tuple from list**


1.   **Mutability:**
*   **Tuple:** Tuples are immutable, meaning once a tuple is created, you cannot modify its elements (add, remove, or change).
*   **List**: Lists are mutable, allowing you to add, remove, or modify elements after the list is created.

2.   **Syntax:**

*   **Tuple:** Created using parentheses ().
*   **List:** Created using square brackets [].

3. **Performance:**

*   Tuples generally have a slight performance advantage over lists due to their immutability.

4. **Use Cases:**


*   **Tuple:** Use when you have a collection of items that should remain constant throughout the program (may contain heterogenous).
*   **List:** Use when you need a collection that may change during the course of the program (contain homogenous elements).

5. **Methods:**

*   **Lists** have more built-in methods compared to tuples, as lists can be modified, appended, and extended.
