**Aim:** Use numpy to perform all possible operations on matrices

```python
import numpy as np
from numpy import linalg as LA
```

```python
mat1 = np.array([[5, 10], [15, 20]])
mat2 = np.array([[1, 3], [5, 4]])
```

**Matrix Addition**

```python
print(np.add(mat1,mat2))
```

**Matrix Subtraction**

```python
print(np.subtract(mat1,mat2))
```

**Matrix Multiplication**

```python
print(np.dot(mat1,mat2))
```

**Summation of elemnts of matrix**

```python
print(np.sum(mat1))
```

**Transpose of a matrix**

```python
print(np.matrix.transpose(mat1))
```

**Retrieve diagonal elements of matrix**

```python
print("Diagonal elements : ",np.diag(mat1))
print("Elements above diagonal : ",np.diag(mat1,1))
print("Elemnts below diagonal : ",np.diag(mat1,-1))
```

**Check if two matrices are equal**

```python
print(np.array_equal(mat1,mat2))
```

**Sort Matrix Elements**

```python
print(np.sort(mat2))
```

**Trace of a matrix**

```python
print(np.trace(mat1))
```

**Heighest and Lowest element in matrix**

```python
print(np.max(mat1))
print(np.min(mat1))
```

**Identity Matrix Creation**

```python
mat3=np.eye(2)
print(mat3)
```

**Determinant of matrix**

```python
print(LA.det(mat1))
```

**Inverse of a matrix**

```python
print(LA.inv(mat1))
```

**Rank of matrix**

```python
print(LA.matrix_rank(mat1))
```

**Eigen Values and Eigen vectors of Matrix**

```python
print(LA.eigvals(mat1))
```

```python
print(LA.eig(mat1))
```
