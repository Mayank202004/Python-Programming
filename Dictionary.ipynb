# **Dictionary in Python**
Dictioanry in python is a data structure that stores the value in value:key pairs.

A dictionary is a collection which is ordered, changeable and do not allow duplicates.

**Creating Dictioanry**

Dictionaries are written with curly brackets {}, and have keys and values:

The elements of dictionary are seperated by comma and key value pair are seperated by colon

```python
dict1={"English":1,"Marathi":2,"Hindi":3,"Science":4,"Social Studies":5}
dict2={1:2,2:4,3:6,4:8,5:10}
type(dict1)
```

**Accessing Dictionary Element**

You can access the items of a dictionary by referring to its key name, inside square brackets:

There is also a method called get() that will give you the same result:

```python
print(dict1["Hindi"]) #value corresponding to "Hindi"
print(dict1.get("Hindi"))
```

**Size of dictionary**

Size of dictionary can be found using len() function

```python
len(dict1)
```

**Adding element in dictioanry**

The addition of elements can be done in multiple ways. One value at a time can be added to a Dictionary by defining value along with the key e.g. Dict[Key] = ‘Value’.

```python
dict1["physics"]=6
print(dict1)
```

**Deleting items in dictionary**

The items of the dictionary can be deleted by using the del keyword as given below.

```python
del(dict1["physics"])
dict1
```

**Check if a key exists in dictionary**

```python
if "Hindi" in dict1:
  print("True")
else:
  print("False")
```

**Dictionary Methods**

**clear() Method**
Remove all the elements from the dictionary

```python
dict2={1:2,2:4,3:6}
dict2.clear()
dict2
```

**copy() Method**

This method returns a copy (shallow copy) of the dictionary.



```python
dict2=dict1.copy()
dict2
```

**copy() method VS = Operator**

When the copy() method is used, a new dictionary is created which is filled with a copy of the references from the original dictionary.
When the = operator is used, a new reference to the original dictionary is created.

With copy() if changes are made in the copy dictionary original dictioanry is not changed and vice versa

with = Operator changes in any dictionary affects each other

```python
dict1={"English":1,"Marathi":2,"Hindi":3,"Science":4,"Social Studies":5}
dict2=dict1.copy()
dict3=dict1
```

```python
dict2['English']=11
print(dict1['English'])
```

```python
dict3['English']=11
print(dict1['English'])
```

```python
#similarly if original dictinary is changed
dict1={"English":1,"Marathi":2,"Hindi":3,"Science":4,"Social Studies":5}
dict2=dict1.copy()
dict3=dict1
dict1["English"]=23
print(dict2["English"])
print(dict3["English"])
```

**items() Method**

Returns a list containing a tuple for each key value pair


```python
dict1.items()
```

**keys() Method**

Returns a list containing dictionary’s keys


```python
dict1.keys()
```

**values() Method**

 Returns a list of all the values of dictionary

```python
dict1.values()
```

**pop() Method**

 Remove the element with specified key

```python
dict1.pop("English")
dict1
```

**popitem() Method**

Removes the last inserted key-value pair

```python
dict1.popitem()
dict1
```

**setdefault Method**

The setdefault() method returns the value of the item with the specified key.

If the key does not exist, insert the key, with the specified value, see example below

```python
#existing key
dict1.setdefault("Science")
```

```python
#Non existing key
dict1.setdefault("Physics")
```
