Importing necessary Libraries

```python
from bs4 import BeautifulSoup
import requests
import html5lib
```

```python
url="https://www.nirfindia.org/2023/EngineeringRanking.html"
```

```python
r=requests.get(url)
soup=BeautifulSoup(r.content,'html5lib')
```

```python
table=soup.find('table')
```

```python
rows=table.find_all('tr')[1:]
rows[6]
```

```python
index =0;
for i in rows[:30:3]:
  column=i.find_all('td')
  name=column[1].text.strip()
  temp=name.split("More Details")
  print(f"{index+1}. {temp[0]}")
  index=index+1
```
