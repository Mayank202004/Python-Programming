**Top Gainers & Losers Stock for the day
Web Scraping using BS4**

import necessary libraies

```python
from bs4 import BeautifulSoup
import requests
import html5lib
```

```python
url = "https://www.moneycontrol.com/stocks/marketstats/nsegainer/index.php"
r=requests.get(url)
soup=BeautifulSoup(r.content,'html5lib')
```

```python
div_with_table = soup.find('div', class_='bsr_table')
table = div_with_table.find('table')
tbody=table.find('tbody')
```

```python
tr=tbody.find_all('tr')
```

```python
print("Top Gainers Today:")
for index,i in enumerate(tr[0:70:7]):
  td=i.find_all('a')
  temp=td[0].text.strip()
  print(f"{index+1}. {temp}")
```
