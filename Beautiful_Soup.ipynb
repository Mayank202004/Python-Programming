```python
import requests
from bs4 import BeautifulSoup
URL = "https://mayank202004.github.io/projects.github.io/visionpro/"
r = requests.get(URL)
soup = BeautifulSoup(r.content, 'html5lib')
table = soup.find('div', attrs = {'id':'page6'})
for row in table:
	print(row)
print("\n\nAfter splitting into dictionary")
page6info = {}
page6info['heading'] = table.find("h1").text
page6info['subheading'] = table.find("h3").text
page6info['paragraph'] = table.find("p").text
print("h1 : ",page6info["heading"])
print("h3 : ",page6info["subheading"])
print("p : ",page6info["paragraph"])
```
