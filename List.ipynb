# **List in Python**
Lists are used to store multiple items in a single variable.

**Creating list**

Lists in Python can be created by just placing the sequence inside the square brackets[].

```python
list1=["English","Marathi","Hindi","Science","Social Studies"]
list2=[1,2,3,4,5]
type(list1)
```

**Accessing List Element**

We can access list elements using list indexing

```python
list1[0] #0'th index element in list1
```

**Reverse Indexing**

Reverse indexing in Python allows you to access elements from the end of a sequence using negative indices.

The last element has an index of -1, the second-to-last has an index of -2, and so on.

```python
print(list1[-1])
print(list2[-4])
```

**List Slicing**

slicing is a technique for extracting a portion (or a subset) of elements of list

The general syntax is **list_name[start:stop:step]**


```python
print(list1[2:4]) #print from 2 upto 4 (excluding)
print(list2[::2]) #print from start to end with step 2
print(list2[::-1]) #print reverse list
```

**Size of List**

Size of list can be found using len() function

```python
len(list1)
```

**String to List**

```python
string1="Hello How are you"
print(type(string1))
list3=string1.split()
print(list3)
print(type(list3))
```

**Append Element in List**

Elements can be added to the List by using the built-in **append()** function. Only one element at a time can be added to the list by using the append() method, for the addition of multiple elements with the append() method, loops are used.

```python
list2=[1,2,3,4,5]
list2.append(6)
list2
```

**Inserting Element at specific position**

**append()** method only works for the addition of elements at the end of the List, for the addition of elements at the desired position, insert() method is used.

**insert()** method requires two arguments(position, value).

```python
list2.insert(1,8) #Insert 8 at index 1
list2
```

**Add Multiple elements at end**

 **extend()** method is used to add multiple elements at the same time at the end of the list.

```python
list2=[1,2,3,4,5]
list2.extend([6,7,8])
print(list2)
list3=[9,10]
#Extend (concatnate)two lists
list2.extend(list3)
print(list2)
```

**Reversing a List**

**reversed()** can be used to reverse a given list.

The reversed() function returns a reverse iterator, which can be converted to a list using the **list()** function.

```python
list2=[1,2,3,4,5]
list3=reversed(list2)
print(list3)
print(list(list3))
```

**remove() Method**

It can be used to remove specific element


```python
list2=[1,2,3,4,5]
list2.remove(2)
list2
```

**pop() Method**

It is used to remove and return an elemnt from list

By default it removes only the last element of the list, to remove a specific element its specific index can be specified

```python
list1=[1,2,3,4,5]
list2=[1,2,3,4,5]
print(list1.pop(2))
print(list1)
print(list2.pop())
print(list1)
```

**index() Method**

Returns index of first matched element

```python
list2=[1,2,3,2,1]
print(list2.index(1)) #returns first matched index of given element
print(list2.index(2))
```

**clear() Method**

Clears entire list (Removes all ements of a list)

```python
list1=[1,2,3]
list1.clear()
list1
```

**count() Method**

It returns the count of the number an items repeats which is passed as an argument

```python
list1=[1,2,3,2,1]
list1.count(2)
```

**sort() Method**



```python
list1=[5,2,8,1,4]
list1.sort() #ascending order
print(list1)
list1.sort(reverse=True) #descending order
print(list1)
```

**copy() Method**

Returns a shallow copy of list

```python
list1=[1,2,3,4,5]
list2=list1.copy()
list2[2]=45
list1[2]
```
